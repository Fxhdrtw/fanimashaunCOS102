```python
import tkinter as tk
from tkinter import messagebox
import random

# Employee Class
class Employee:
    def __init__(self):
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali",
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Anide",
            "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]

        self.tasks = [
            "Loading", "Transporting", "Reviewing Orders",
            "Customer Service", "Delivering Items"
        ]

    def check_employee(self, name):
        return name in self.employees

    def take_attendance(self, name):
        return f"Attendance taken for {name}"

    def assign_task(self):
        return random.choice(self.tasks)

    def refuse_access(self):
        return "Access Denied! You are not an employee."


# GUI Class
class App:
    def __init__(self, root):
        self.employee = Employee()
        self.root = root
        self.root.title("Employee System")

        # Label
        self.label = tk.Label(root, text="Enter Employee Name:")
        self.label.pack(pady=10)

        # Entry box
        self.entry = tk.Entry(root, width=30)
        self.entry.pack(pady=5)

        # Button
        self.button = tk.Button(root, text="Submit", command=self.process)
        self.button.pack(pady=10)

        # Output
        self.output = tk.Label(root, text="", fg="blue")
        self.output.pack(pady=10)

    def process(self):
        name = self.entry.get()

        if self.employee.check_employee(name):
            attendance = self.employee.take_attendance(name)
            task = self.employee.assign_task()

            result = f"{attendance}\nTask Assigned: {task}"
        else:
            result = self.employee.refuse_access()

        self.output.config(text=result)


# Run App
root = tk.Tk()
app = App(root)
root.mainloop()
```
