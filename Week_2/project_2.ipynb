```python
# Polynomial Root Finder

print("Root Finder")
print("1. Quadratic Equation (Ax^2 + Bx + C = 0)")
print("2. Cubic Equation (Ax^3 + Bx^2 + Cx + D = 0)")
print("3. Quartic Equation (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")

choice = int(input("Select the equation type (1/2/3): "))

# Quadratic Equation
if choice == 1:
    print("\nQuadratic Equation: Ax^2 + Bx + C = 0")
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))

    D = B**2 - 4*A*C

    if D > 0:
        r1 = (-B + D**0.5) / (2*A)
        r2 = (-B - D**0.5) / (2*A)
        print("Roots are:", r1, "and", r2)

    elif D == 0:
        r = -B / (2*A)
        print("Both roots are:", r)

    else:
        real = -B / (2*A)
        imag = (abs(D)**0.5) / (2*A)
        print("Roots are complex:", real, "+", imag, "i and", real, "-", imag, "i")


# Cubic Equation (simple numerical search)
elif choice == 2:
    print("\nCubic Equation: Ax^3 + Bx^2 + Cx + D = 0")
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))

    print("Approximate real roots:")
    for x in range(-100, 101):
        value = A*x**3 + B*x**2 + C*x + D
        if abs(value) < 0.01:
            print("Root ≈", x)


# Quartic Equation (simple numerical search)
elif choice == 3:
    print("\nQuartic Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0")
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    E = float(input("Enter E: "))

    print("Approximate real roots:")
    for x in range(-100, 101):
        value = A*x**4 + B*x**3 + C*x**2 + D*x + E
        if abs(value) < 0.01:
            print("Root ≈", x)

else:
    print("Invalid choice")
```
