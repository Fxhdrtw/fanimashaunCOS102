```python
# Program to take two names and their ages and swap the ages

# Taking input from the user
name1 = input("Enter the first name: ")
age1 = int(input(f"Enter the age of {name1}: "))

name2 = input("Enter the second name: ")
age2 = int(input(f"Enter the age of {name2}: "))

# Swapping the ages
temp = age1
age1 = age2
age2 = temp

# Displaying the results
print("\nAfter swapping ages:")
print(name1, "is now", age1, "years old")
print(name2, "is now", age2, "years old")
```
