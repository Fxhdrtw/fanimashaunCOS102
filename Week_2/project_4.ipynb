```python
# Program to determine Annual Tax Revenue (ATR)

print("IZIFIN TECHNOLOGY - STAFF ATR CALCULATOR")

# Input from user
experience = int(input("Enter years of work experience: "))
age = int(input("Enter staff age: "))

# Decision structure
if experience > 25 and age >= 55:
    atr = 5600000

elif experience > 20 and age >= 45:
    atr = 4480000

elif experience > 10 and age >= 35:
    atr = 1500000

else:
    atr = 550000

# Output result
print("Annual Tax Revenue (ATR) is: N", atr)
```
