```python
# Simple Interest
P = float(input("Enter Principal: "))
R = float(input("Enter Rate (%): "))
T = float(input("Enter Time (years): "))

SI = (P * R * T) / 100

print("Simple Interest =", SI)
```

```python
# Compound Interest
P = float(input("Enter Principal: "))
R = float(input("Enter annual interest rate (decimal): "))
n = int(input("Enter number of times interest applied per year: "))
t = float(input("Enter time in years: "))

A = P * (1 + R/n) ** (n*t)

print("Total Amount =", A)
print("Compound Interest =", A - P)
```

```python
# Annuity Plan 
PMT = float(input("Enter payment per period: "))
R = float(input("Enter annual interest rate (decimal): "))
n = int(input("Enter number of compounding periods per year: "))
t = float(input("Enter number of years: "))

A = PMT * ((1 + R/n) ** (n*t) - 1) / (R/n)

print("Future Value of Annuity =", A)
```
