```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load dataset
data = pd.read_excel("GIG-logistics.xlsx")

def check_employee():
    name = name_entry.get().strip()
    dept = dept_entry.get().strip()

    # Filter dataset
    result = data[
        (data["Name"].str.lower() == name.lower()) &
        (data["Department"].str.lower() == dept.lower())
    ]

    if not result.empty:
        # Employee exists
        same_dept = data[data["Department"].str.lower() == dept.lower()]

        others = same_dept["Name"].tolist()

        messagebox.showinfo(
            "Success",
            f"Welcome {name}!\n\nOther employees in {dept}:\n" + "\n".join(others)
        )
    else:
        messagebox.showerror("Error", "Employee does not exist")

# GUI
root = tk.Tk()
root.title("Employee Verification")
root.geometry("400x200")

tk.Label(root, text="Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Department").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Check", command=check_employee).pack()

root.mainloop()
```
